## Creating simple plots using nextGEMS cycle 3 data

The data analized here are:
- tco1279-orca025 (5 years)
- tco2559-ng5 (5 years)
- ngc3028 (5 years)

In this notebook we will have a look at the variables and have a look at some relevant parameters that should remain constant along a climatology (e.g. the 2m temperature) Different variables are chosen for the comparison.

```python
from aqua import Reader
import sys
sys.path.insert(0, '../../')
from atmglobalmean import plot_map_with_stats
```

The datasets we are having a look at are:

```python
reader_era5 = Reader(model="ERA5", exp="era5", source="monthly")
data_era5 = reader_era5.retrieve()

reader_tco2559 = Reader(model = 'IFS', exp = 'tco2559-ng5-cycle3', source = 'lra-r100-monthly')
data_tco2559 = reader_tco2559.retrieve()

reader_tco1279 = Reader(model="IFS", exp="tco1279-orca025-cycle3",source =  'lra-r100-monthly')
data_tco1279 = reader_tco1279.retrieve()

reader_icon = Reader(model = "ICON", exp = "ngc3028", source = 'lra-r100-monthly')
data_icon = reader_icon.retrieve()
```

#### Now we can plot a required variable from a required dataset and show the statistics:

```python
plot_map_with_stats(dataset=data_tco2559, var_name='2t', start_date='2021-01-01', end_date='2023-12-31', model_label='IFS4.4km')
plot_map_with_stats(dataset=data_tco1279, var_name='2t', start_date='2021-01-01', end_date='2023-12-31', model_label='IFS9km')
plot_map_with_stats(dataset=data_icon, var_name='2t', start_date='2021-01-01', end_date='2023-12-31', model_label='ICON5km')
```

```python
plot_map_with_stats(dataset=data_tco2559, var_name='msl', start_date='2020-01-01', end_date='2024-12-31', model_label='IFS4.4km')
plot_map_with_stats(dataset=data_tco1279, var_name='msl', start_date='2020-01-01', end_date='2024-12-31', model_label='IFS 9 km')
```
