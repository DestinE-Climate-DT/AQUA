## Creating zonal bias plots using nextGEMS cycle 3 data

The data analized here are:
- tco1279-orca025 (5 years)
- tco2559-ng5 (5 years)
- ngc3028 (5 years)

In this notebook we will have a look at the vertical bias plots where we compare the model data with ERA5 climatology. The plots show meridional cross sections.
Different variables are chosen for the comparison.

```python
from aqua import Reader
import sys
sys.path.insert(0, '../../')
from atmglobalmean import compare_datasets_plev
```

The datasets we are having a look at are:

```python
reader_era5 = Reader(model="ERA5", exp="era5", source="monthly")
data_era5 = reader_era5.retrieve()

reader_tco2559 = Reader(model = 'IFS', exp = 'tco2559-ng5-cycle3', source = 'lra-r100-monthly')
data_tco2559 = reader_tco2559.retrieve()

reader_tco1279 = Reader(model="IFS", exp="tco1279-orca025-cycle3",source =  'lra-r100-monthly')
data_tco1279 = reader_tco1279.retrieve()

reader_icon = Reader(model = "ICON", exp = "ngc3028", source = 'lra-r100-monthly')
data_icon = reader_icon.retrieve()
```

```python
# Specify the dataset1, dataset2, var_name, plev, statistic, model labels, and climatology ranges
dataset1 = data_tco2559
dataset2 = data_era5
start_date1 = '2020-01-01'
end_date1 = '2024-12-31'
start_date2 = '2000-01-01'
end_date2 = '2020-12-31'
var_name = 't'
model_label1 = 'IFS_4km'
model_label2 = 'ERA5'

compare_datasets_plev(dataset1=dataset1, dataset2=dataset2, var_name=var_name, start_date1=start_date1, end_date1=end_date1, 
                      start_date2=start_date2, end_date2=end_date2, model_label1=model_label1, model_label2=model_label2)
```
