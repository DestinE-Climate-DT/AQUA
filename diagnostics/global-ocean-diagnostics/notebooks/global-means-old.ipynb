```python
t0=t3d[:,0,:,:].mean(("i","j"))
t100=t3d[:,22,:,:].mean(("i","j"))
t500=t3d[:,39,:,:].mean(("i","j"))
t1000=t3d[:,46,:,:].mean(("i","j"))
t2000=t3d[:,53,:,:].mean(("i","j"))
t3000=t3d[:,59,:,:].mean(("i","j"))
t4000=t3d[:,64,:,:].mean(("i","j"))

# To compute the temporal anomalies
t0a=t0-t0.mean("time")
t100a=t100-t100.mean("time")
t500a=t500-t500.mean("time")
t1000a=t1000-t1000.mean("time")
t2000a=t2000-t2000.mean("time")
t3000a=t3000-t3000.mean("time")
t4000a=t4000-t4000.mean("time")
```

```python
tu0=t3dunweighted[:,0,:,:].mean(("i","j"))
tu100=t3dunweighted[:,22,:,:].mean(("i","j"))
tu500=t3dunweighted[:,39,:,:].mean(("i","j"))
tu1000=t3dunweighted[:,46,:,:].mean(("i","j"))
tu2000=t3dunweighted[:,53,:,:].mean(("i","j"))
tu3000=t3dunweighted[:,59,:,:].mean(("i","j"))
tu4000=t3dunweighted[:,64,:,:].mean(("i","j"))

# To compute the temporal anomalies
tu0a=tu0-tu0.mean("time")
tu100a=tu100-tu100.mean("time")
tu500a=tu500-tu500.mean("time")
tu1000a=tu1000-tu1000.mean("time")
tu2000a=tu2000-tu2000.mean("time")
tu3000a=tu3000-tu3000.mean("time")
tu4000a=tu4000-tu4000.mean("time")
```

```python
t0sm=t0.rolling(time=12,center=True).mean("time")
t100sm=t100.rolling(time=12,center=True).mean("time")
t500sm=t500.rolling(time=12,center=True).mean("time")
t1000sm=t1000.rolling(time=12,center=True).mean("time")
t2000sm=t2000.rolling(time=12,center=True).mean("time")
t3000sm=t3000.rolling(time=12,center=True).mean("time")

t0sm.plot.line()
t100sm.plot.line()
t500sm.plot.line()
t1000sm.plot.line()
t2000sm.plot.line()
t3000sm.plot.line()

#t0.plot.line(color="blue",linestyle="dotted")
plt.title("Evolution of the global ocean mean temperature")
plt.legend(["0","100","500","1000","2000","3000"], loc='best')
#t100sm.plot.line(color="green")
#plt.legend(["0m","100m"])
```

```python
# We apply a 12 month running mean to smooth the data to exhibit only annual variations
t0asm=t0a.rolling(time=12,center=True).mean("time")
t100asm=t100a.rolling(time=12,center=True).mean("time")
t500asm=t500a.rolling(time=12,center=True).mean("time")
t1000asm=t1000a.rolling(time=12,center=True).mean("time")
t2000asm=t2000a.rolling(time=12,center=True).mean("time")
t3000asm=t3000a.rolling(time=12,center=True).mean("time")
t4000asm=t3000a.rolling(time=12,center=True).mean("time")

t0asm.plot.line()
t100asm.plot.line()
t500asm.plot.line()
t1000asm.plot.line()
t2000asm.plot.line()
t3000asm.plot.line()
t4000asm.plot.line()

#t0.plot.line(color="blue",linestyle="dotted")
plt.title("Evolution of the global ocean mean temperature anomaly")
plt.legend(["0","100","500","1000","2000","3000","4000"], loc='best')
#t100sm.plot.line(color="green")
#plt.legend(["0m","100m"])
```

```python
# We standardise the data to ease the comparison between levels
t0ssm=t0asm/t0asm.std()
t100ssm=t100asm/t100asm.std()
t500ssm=t500asm/t500asm.std()
t1000ssm=t1000asm/t1000asm.std()
t2000ssm=t2000asm/t2000asm.std()
t3000ssm=t3000asm/t3000asm.std()
t4000ssm=t4000asm/t4000asm.std()

#t100sm.plot.line(color="green")
#plt.legend(["0m","100m"])
t0ssm.plot.line()
t100ssm.plot.line()
t500ssm.plot.line()
t1000ssm.plot.line()
t2000ssm.plot.line()
t3000ssm.plot.line()
t4000ssm.plot.line()
plt.title("Evolution of the global ocean mean temperature anomaly (standardised)")
plt.legend(["0","100","500","1000","2000","3000","4000"], loc='best')
```
