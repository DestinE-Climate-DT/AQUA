```python
import os
import xarray as xr
import matplotlib.pyplot as plt
```

```python
# Open the file
outputdir='output/netcdf'
icon_histo_ts_mon_file=os.path.join(outputdir,
                                    'global_time_series_timeseries_2t_ICON_historical-1990_mon.nc')
icon_histo_ts_mon=xr.open_mfdataset(icon_histo_ts_mon_file)
icon_histo_ts_ann=os.path.join(outputdir,
                                 'global_time_series_timeseries_2t_ICON_historical-1990_ann.nc')
icon_histo_ts_ann=xr.open_mfdataset(icon_histo_ts_ann)
icon_ssp_ts_mon_file=os.path.join(outputdir,
                                  'global_time_series_timeseries_2t_ICON_ssp370_mon.nc')
icon_ssp_ts_mon=xr.open_mfdataset(icon_ssp_ts_mon_file)
icon_ssp_ts_ann=os.path.join(outputdir,
                                 'global_time_series_timeseries_2t_ICON_ssp370_ann.nc')
icon_ssp_ts_ann=xr.open_mfdataset(icon_ssp_ts_ann)
```

```python
fig = plt.figure(figsize=(10, 10))

# Top line, gregory plot
ax1 = plt.subplot(2, 2, 1)

# Top line, seasonal cycle
ax2 = plt.subplot(2, 2, 2)

# Bottom line, timeseries
ax3 = plt.subplot(2, 1, 2)
icon_histo_ts_mon['2t'].plot(ax=ax3)
```
