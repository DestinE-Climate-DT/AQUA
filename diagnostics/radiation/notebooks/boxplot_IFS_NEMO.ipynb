### Creating a simple boxplot to show imbalances of the models

```python
import sys
sys.path.insert(0,'../../')
from radiation import process_model_data, process_ceres_data
from radiation import boxplot_model_data
```

### Loading the data we need for the analysis

```python
ceres_toa = process_ceres_data(exp = "ebaf-toa41", source = "monthly", level = 'toa', fix=True)
ifs_nemo_historical = process_model_data(model =  'IFS-NEMO' , exp = 'historical-1990' , source = 'lra-r100-monthly', fix = True)
era5 = process_model_data(model = 'ERA5', exp = 'era5', source = 'monthly', fix = True)
ceres_sfc = process_ceres_data(exp = "ebaf-sfc41", source = "monthly", level = 'sfc', fix=True)
```

```python
datasets = [ceres_toa, ifs_nemo_historical, era5]
variables = ['-mtnlwrf', 'mtnswrf']

boxplot_model_data(datasets=datasets,variables=variables)
```

```python
datasets = [ceres_sfc, ifs_nemo_historical, era5]
variables = ['msnswrf']

boxplot_model_data(datasets=datasets,variables=variables)
```
