# ENSO control run

## Running ENSO teleconnection

```python
from teleconnections import Teleconnection
```

```python
telecname = 'ENSO_2t'

model = 'IFS'
exp = 'control-1950-devcon'
source = 'lra-r100-monthly'
```

```python
telec = Teleconnection(model=model, exp=exp, source=source, telecname=telecname)
```

## Index

```python
telec.evaluate_index()
```

```python
telec.plot_index()
```

## Regression

```python
from teleconnections.plots import plot_single_map

reg = telec.evaluate_regression()

plot_single_map(data=reg, cbar_label='2t [K]', title='ENSO regression of IFS control-1950-devcon')
```

## Correlation

```python
cor = telec.evaluate_correlation()

plot_single_map(data=cor, cbar_label='Pearson coefficient', title='ENSO correlation of IFS control-1950-devcon')
```
