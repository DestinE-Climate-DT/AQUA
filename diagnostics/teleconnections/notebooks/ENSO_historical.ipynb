```python
from teleconnections import Teleconnection
```

```python
tc_ERA = Teleconnection(model='ERA5', exp='era5', source='monthly', loglevel='DEBUG', telecname='ENSO', startdate="1990-01-01", enddate="1998-11-01", savefig=True)
```

```python
tc_ERA.evaluate_index()
```

```python
tc_ERA.plot_index(title='ENSO index (ERA5, 1990-1998)')
```

```python
reg_ERA = tc_ERA.evaluate_regression()
cor_ERA = tc_ERA.evaluate_correlation()
reg_ERA_tprate = tc_ERA.evaluate_regression(var='tprate')
```

```python
tc_IFS = Teleconnection(model='IFS-NEMO', exp='historical-1990', source='lra-r100-monthly', loglevel='INFO', telecname='ENSO', savefig=True, enddate="1998-11-01")
```

```python
tc_IFS.evaluate_index()
```

```python
tc_IFS.plot_index(title='ENSO index (IFS-NEMO, 1990-1998)')
```

```python
reg_IFS = tc_IFS.evaluate_regression()
cor_IFS = tc_IFS.evaluate_correlation()
reg_IFS_tprate = tc_IFS.evaluate_regression(var='tprate')
```

```python
reg_IFS_tprate.aqua.plot_single_map()
```

```python
from teleconnections import maps_plot
```

```python
maps_reg = [reg_ERA, reg_IFS, reg_ERA_tprate, reg_IFS_tprate]
titles = ['ERA5 regression (avg_tos)', 'IFS-NEMO regression (avg_tos)', 'ERA5 regression (tprate)', 'IFS-NEMO regression (tprate)']
```

```python
maps_plot(maps_reg, title='ENSO regression maps (1990-1998)', titles=titles, save=True, filename='ENSO_regression_maps.pdf')
```

```python
cor_maps = [cor_ERA, cor_IFS]
titles = ['ERA5 correlation', 'IFS-NEMO correlation']

maps_plot(cor_maps, title='ENSO correlation maps (1990-1998)', titles=titles, save=True, filename='ENSO_correlation_maps.pdf')
```
