```python
tc_ERA = Teleconnection(model='ERA5', exp='era5', source='monthly', loglevel='DEBUG', telecname='ENSO', startdate="1990-01-01", enddate="1998-11-01", savefig=True)
```

```python
tc_ERA.evaluate_index()
```

```python
tc_ERA.plot_index(title='ENSO index (ERA5, 1990-1998)', ylim=(-2.5,2.5))
```

```python
reg_ERA = tc_ERA.evaluate_regression()
cor_ERA = tc_ERA.evaluate_correlation()
```

```python
tc_IFS = Teleconnection(model='IFS-NEMO', exp='historical-1990', source='lra-r100-monthly', loglevel='INFO', telecname='ENSO', savefig=True, enddate="1998-11-01")
```

```python
tc_IFS.evaluate_index()
```

```python
tc_IFS.plot_index(title='ENSO index (IFS-NEMO, 1990-1998)', ylim=(-2.5,2.5))
```

```python
reg_IFS = tc_IFS.evaluate_regression()
cor_IFS = tc_IFS.evaluate_correlation()
```

```python
from aqua.util import evaluate_colorbar_limits
vmin, vmax = evaluate_colorbar_limits([reg_ERA, reg_IFS], sym=True)
print(vmin, vmax)
```

```python
reg_ERA.aqua.plot_single_map(title='ENSO regression (ERA5, 1990-1998)', vmin=vmin, vmax=vmax, cmap='RdBu_r', save=True,
                             filename='ENSO_regression_ERA5_1990-1998.pdf')
```

```python
reg_IFS.aqua.plot_single_map(title='ENSO regression (IFS-NEMO, 1990-1998)', vmin=vmin, vmax=vmax, cmap='RdBu_r', save=True,
                                filename='ENSO_regression_IFS-NEMO_1990-1998.pdf', transform_first=True)
```

```python
vmin, vmax = evaluate_colorbar_limits([cor_ERA, cor_IFS], sym=True)
print(vmin, vmax)
```

```python
cor_IFS.aqua.plot_single_map(title='ENSO correlation (IFS-NEMO, 1990-1998)', vmin=vmin, vmax=vmax, cmap='RdBu_r', save=True,
                             filename='ENSO_correlation_IFS-NEMO_1990-1998.pdf', transform_first=True)
```

```python
cor_ERA.aqua.plot_single_map(title='ENSO correlation (ERA5, 1990-1998)', vmin=vmin, vmax=vmax, cmap='RdBu_r', save=True,
                                filename='ENSO_correlation_ERA5_1990-1998.pdf')
```
