## Setting up the teleconnection class

```python
from teleconnections import Teleconnection
```

```python
telecname = 'NAO'

model = 'IFS'
exp = 'control-1950-devcon'
source = 'lra-r100-monthly'
```

### Initialize the class

```python
telec = Teleconnection(model=model, exp=exp, source=source, telecname=telecname)
```

Notice that warning about regridding and time aggregation are printed out, since we are not specifying any of them, but we're using an LRA already time aggregated and regridded.

## Runnning the teleconnection

```python
telec.run()
```

## NAO index

```python
telec.plot_index()
```

## Regression

```python
from teleconnections.plots import single_map_plot

single_map_plot(map=telec.regression, model=telec.model, exp=telec.exp, cbar_label='msl (Pa)', title='NAO regression of IFS control-1950-devcon')
```

## Correlation


```python
single_map_plot(map=telec.correlation, model=telec.model, exp=telec.exp, cbar_label='Pearson coefficient', title='NAO correlation of IFS control-1950-devcon')
```
