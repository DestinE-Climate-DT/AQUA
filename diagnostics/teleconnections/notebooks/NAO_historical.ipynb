```python
from teleconnections import Teleconnection
```

```python
tc_ERA = Teleconnection(model='ERA5', exp='era5', source='monthly', loglevel='INFO', telecname='NAO', startdate="1990-01-01", enddate="1998-11-01", savefig=True)
```

```python
tc_ERA.evaluate_index()
```

```python
tc_ERA.plot_index(title='NAO index (ERA5, 1990-1998)', ylim=(-3.5, 3.5))
```

```python
reg_ERA_DFJ = tc_ERA.evaluate_regression(season='DJF')
reg_ERA_JJA = tc_ERA.evaluate_regression(season='JJA')
cor_ERA_DJF = tc_ERA.evaluate_correlation(season='DJF')
cor_ERA_JJA = tc_ERA.evaluate_correlation(season='JJA')
```

```python
tc_IFS = Teleconnection(model='IFS-NEMO', exp='historical-1990', source='lra-r100-monthly', loglevel='INFO', telecname='NAO', savefig=True)
```

```python
tc_IFS.evaluate_index()
```

```python
tc_IFS.plot_index(title='NAO index (IFS-NEMO, 1990-1998)', ylim=(-3.5, 3.5))
```

```python
reg_IFS_DFJ = tc_IFS.evaluate_regression(season='DJF')
reg_IFS_JJA = tc_IFS.evaluate_regression(season='JJA')
cor_IFS_DJF = tc_IFS.evaluate_correlation(season='DJF')
cor_IFS_JJA = tc_IFS.evaluate_correlation(season='JJA')
```

```python
from teleconnections import maps_plot
```

```python
maps_reg_DJF = [reg_ERA_DFJ, reg_IFS_DFJ]
maps_reg_JJA = [reg_ERA_JJA, reg_IFS_JJA]
titles_reg_DJF = ['ERA5 DJF regression', 'IFS-NEMO DJF regression']
titles_reg_JJA = ['ERA5 JJA regression', 'IFS-NEMO JJA regression']
```

```python
%reload_ext autoreload
%autoreload 2
```

```python
maps_plot(maps_reg_DJF, title='NAO regression maps DJF (1990-1998)', titles=titles_reg_DJF,
          save=True, filename='NAO_regression_maps_DJF.pdf', nlevels=11, figsize=(11, 5), cbar_label='msl (Pa)')
```

```python
maps_plot(maps_reg_JJA, title='NAO regression maps JJA (1990-1998)', titles=titles_reg_JJA,
          save=True, filename='NAO_regression_maps_JJA.pdf', nlevels=11, figsize=(11, 5), cbar_label='msl (Pa)')
```

```python
cor_maps_DJF = [cor_ERA_DJF, cor_IFS_DJF]
cor_maps_JJA = [cor_ERA_JJA, cor_IFS_JJA]
titles_DJF = ['ERA5 DJF correlation', 'IFS-NEMO DJF correlation']
titles_JJA = ['ERA5 JJA correlation', 'IFS-NEMO JJA correlation']
```

```python
maps_plot(cor_maps_DJF, title='NAO correlation maps DJF (1990-1998)', titles=titles_DJF, vmin=-1, vmax=1,
            save=True, filename='NAO_correlation_maps_DJF.pdf', nlevels=11, figsize=(11, 5), cbar_label='correlation coefficient')
```

```python
maps_plot(cor_maps_JJA, title='NAO correlation maps JJA (1990-1998)', titles=titles_JJA, vmin=-1, vmax=1,
            save=True, filename='NAO_correlation_maps_JJA.pdf', nlevels=11, figsize=(11, 5), cbar_label='correlation coefficient')
```
