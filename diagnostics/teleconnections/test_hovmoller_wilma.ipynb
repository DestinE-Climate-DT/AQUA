```python
import sys
from plots import *
from tools import *
```

```python
config = load_config()
namelist = load_namelist()
```

```python
telecname = 'MJO'
fieldname = namelist[telecname]['field']
infile = config['teleconnections'][telecname]['input']
field = xr.open_dataset(infile)[fieldname]
xlim = ['1995-01-01T06:00:00.000000000', '1996-01-01T06:00:00.000000000']
ylim = [namelist[telecname]['latN'], namelist[telecname]['latS']]
field = field.sel(time=slice(xlim[0], xlim[1]))
field = field.sel(lat=slice(ylim[0], ylim[1]))
field = field - field.rolling(time=3, center=True).mean()
```

```python
hovmoller_plot(field,dim='lat',contour=True,levels=15,title='MJO',invert_axis=True)
```
