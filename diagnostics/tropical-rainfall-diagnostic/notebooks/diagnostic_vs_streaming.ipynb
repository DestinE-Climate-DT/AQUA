## 1. The load of packages, modules, and data

```python
from aqua import Reader
import sys
sys.path.insert(0, '../')
from tropical_rainfall_class import TR_PR_Diagnostic as TR_PR_Diag
```

```python
diag = TR_PR_Diag(num_of_bins = 20, first_edge = 0, width_of_bin = 1*10**(-6)/15)
```

```python
reader = Reader(model="ICON", exp="ngc2009", source="atm_2d_ml_R02B09",  regrid="r100")
```

## 2. Calculating and saving the histograms in the storage for each stream output (ten in total)

```python
path_to_save='/work/bb1153/b382267/AQUA/histograms/'

for i in range(0, 10):
    icon_3_days = reader.retrieve(regrid=True, streaming=True, stream_step=3, stream_unit='days') 
    hist_icon  = diag.histogram(icon_3_days)
    diag.save_histogram(dataset=hist_icon, path_to_save=path_to_save, name_of_file='icon')
```

## 3. Loading and merging all the histograms in the specified repository 

```python
path_to_save='/work/bb1153/b382267/AQUA/histograms/'
```

```python
merged_histograms = diag.merge_list_of_histograms(path_to_histograms=path_to_save, multi=None, all=True)
merged_histograms
```

## 4. Loading and merging multiple histograms in the specified repository 

```python
merged_histograms = diag.merge_list_of_histograms(path_to_histograms=path_to_save, multi=3, all=False)
merged_histograms
```
