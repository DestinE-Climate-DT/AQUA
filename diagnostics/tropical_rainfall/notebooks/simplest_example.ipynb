```python
from aqua import Reader, catalogue
import sys
sys.path.insert(0, '../')
from src.trop_prec_diagnostic import TR_PR_Diagnostic as TR_PR_Diag
```

```python
reader = Reader(model="IFS", exp="tco2559-ng5", source="ICMGG_atm2d", regrid="r010")
ifs = reader.retrieve(regrid=True)
```

```python
diag = TR_PR_Diag(trop_lat = 10,  num_of_bins = 15, first_edge = 0, width_of_bin = 1*10**(-6)/15)
```

## To make the test very fast, loading only one timestep

```python
chunk_1  = reader.retrieve(streaming=True, stream_step =1, regrid=True )
chunk_2  = reader.retrieve(streaming=True, stream_step =1, regrid=True)
chunk_3  = reader.retrieve(streaming=True, stream_step =1, regrid=True)
```

##  Chunk #1

#### Counts-histogram of tropical precipitaion

```python

hist_ifs_week_1  = diag.hist1d_fast(chunk_1)
hist_ifs_week_1
```

#### Creating the Dataset with global attributes

```python
tprate_dataset_1 = hist_ifs_week_1.to_dataset(name="trop_counts")
tprate_dataset_1.attrs = ifs.attrs
```

#### Storing into memory

```python
path = '.'
with open(str(path)+'ifs_dataset_1.pickle', 'wb') as output:
    pickle.dump(tprate_dataset_1, output)
```
