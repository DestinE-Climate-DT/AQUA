```python
from aqua import Reader
from ocean3d.ocean_drifts.trends import TrendCalculator
```

```python
reader = Reader(model='ICON', exp='ssp370', source="lra-r100-monthly")
data = reader.retrieve(var="2t")
```

```python
trend = TrendCalculator.TS_3dtrend(data)
trend
```

```python
trend['2t'].aqua.plot_single_map(
    contour=False,
    # save=True,
    loglevel="info",
    format="png" , 
    dpi=300, 
    model="ICON",
    exp="SSP370"
    )
```
