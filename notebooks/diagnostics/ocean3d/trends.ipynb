```python
from aqua.diagnostics.ocean_trends import Trends
from aqua.diagnostics.ocean_trends.plot_trends import PlotTrends
```

### Initalizing the function

```python
trend = Trends(catalog='mn5-phase2', model='IFS-NEMO', exp='control-1990-freeze', source='lra-r100-monthly',
                startdate="01-01-1990", enddate="31-12-1990",
          loglevel='DEBUG')
```

```python
trend.run(
    outputdir= ".",
    var=['thetao', 'so'],
    region="io")
```

### Plotting the data

```python
ml_trend = PlotTrends(data = trend.trend_coef)
ml_trend.plot_multilevel()
```

```python
zm_trend = PlotTrends(data = trend.trend_coef.mean('lon'))
zm_trend.plot_zonal()
```
