```python
from aqua.diagnostics.ocean_trends import Trends
```

```python
trend = Trends(
    # loglevel='DEBUG',
    loglevel='INFO',
    catalog='ci',
    exp='hpz3',
    model='FESOM',
    source='monthly-3d',
    regrid='r100'
)
```

```python
trend.run(dim_mean="lat",
    outputdir= ".",
    var=['thetao', 'so'],
    region="io")
```

```python
trend.trend_coef.thetao.plot()
```
