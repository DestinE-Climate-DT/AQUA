```python
from aqua.diagnostics.ocean_drift.hovmoller import Hovmoller
from aqua.diagnostics.ocean_drift.plot_hovmoller import PlotHovmoller
```

```python
hov = Hovmoller(
    catalog='mn5-phase2',
    model='IFS-NEMO',
    exp='control-1990-freeze',
    source='lra-r100-monthly',
    startdate='01-01-1991',
    enddate='31-05-1991',
    loglevel='INFO'
)
```

```python
hov.run(
    region='io',
    dim_mean=['lat', 'lon'],
    anomaly_ref='t0'
)
```

```python
hov_plot = PlotHovmoller(data = hov.processed_data_list)
hov_plot.plot_hovmoller()
```

```python
hov_plot.plot_timeseries()
```
