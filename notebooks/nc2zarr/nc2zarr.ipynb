## Plots

#### Timing

```python
fig, ax =  plt.subplots( ) 
ax.plot([1, test_number_of_files, total_number_of_files],
        [time_netcdf_av/test_number_of_files, time_netcdf_av, total_number_of_files*time_netcdf_av/test_number_of_files], 
        label= 'MultiNetCDF', lw=3, color = 'tab:blue')

ax.plot([1, test_number_of_files, total_number_of_files],
        [time_zarr_av/test_number_of_files, time_zarr_av, total_number_of_files*time_zarr_av/test_number_of_files], 
        label= 'MultiZarr', lw=3, color = 'tab:red')

ax.plot([1, test_number_of_files, total_number_of_files],
        [time_SZarr_av/test_number_of_files, time_SZarr_av, total_number_of_files*time_SZarr_av/test_number_of_files], 
        label= 'SingleZarr', lw=3, color = 'tab:green')
#seconds2hours = 1/(60*60)
sec2min = 1/(60)
ax2=ax.twinx()
ax2.plot([1, test_number_of_files, total_number_of_files],
         [time_zarr_av*sec2min /test_number_of_files, time_zarr_av*sec2min, total_number_of_files*time_zarr_av*sec2min /test_number_of_files], 
         ls=' ')
ax.set_yscale('log')
ax2.set_yscale('log')
ax.set_xscale('log')
ax2.set_xscale('log')
ax.set_ylabel('Seconds', fontsize=12)
ax2.set_ylabel('Minutes', fontsize=12)
ax.set_title('Time to read the metadata', fontsize=14)

ax.grid(True)
ax2.grid(True)
ax.legend(fontsize = 14)
```

#### Memory Usage

```python

max_number_of_files =  total_number_of_files

fig, ax =  plt.subplots( ) 
# NetCDF
mem_size_of_cdf_file, mem_unit_of_cdf_file = mem_usage.mem_of_file(single_cdf_file)
mem_cdf_converter = mem_usage.mem_unit_converter(old_unit=mem_unit_of_cdf_file, new_unit='G')
ax.plot([1, max_number_of_files],
        [mem_size_of_cdf_file, max_number_of_files*mem_size_of_cdf_file], 
        label= 'MultiNetCDF', lw=3, color = 'tab:blue')
# Zarr

mem_size_of_zarr_file, mem_unit_of_zarr_file = mem_usage.mem_of_file(single_zarr_file)
mem_zarr_converter = mem_usage.mem_unit_converter(old_unit=mem_unit_of_zarr_file, new_unit=mem_unit_of_cdf_file)
ax.plot([1, max_number_of_files],
        [mem_size_of_zarr_file*mem_zarr_converter, max_number_of_files*mem_size_of_zarr_file*mem_zarr_converter ], 
        label= 'MultiZarr', lw=3, color = 'tab:purple')

mem_size_of_SZarr_file, mem_unit_of_SZarr_file = mem_usage.mem_of_file(singleZpath)
mem_SZarr_converter = mem_usage.mem_unit_converter(old_unit = mem_unit_of_SZarr_file, new_unit=mem_unit_of_cdf_file)
ax.plot([1, max_number_of_files],
        [mem_size_of_SZarr_file*mem_SZarr_converter, max_number_of_files*mem_size_of_SZarr_file*mem_SZarr_converter ], 
        label= 'SingleZarr', lw=3, color = 'k')

ax2=ax.twinx()
ax2.plot([1, max_number_of_files],
        [mem_size_of_cdf_file*mem_cdf_converter, max_number_of_files*mem_size_of_cdf_file*mem_cdf_converter],
         ls=' ')
ax.set_yscale('log')
ax2.set_yscale('log')
ax.set_xscale('log')
ax2.set_xscale('log')
ax.set_ylabel(mem_unit_of_cdf_file, fontsize=12)
ax2.set_ylabel('G', fontsize=12)




# compute 
max_memory, max_walltime, max_cpus, max_sockets, max_cores, max_threads  = mem_usage.max_resources_per_node("compute")
max_memory_compute = max_memory

mem_size, mem_unit = mem_usage.exctract_mem_size_and_unit(max_memory_compute)
mem_converter = mem_usage.mem_unit_converter(old_unit=mem_unit, new_unit='G')
ax2.axhline(mem_converter*mem_size, color = 'tab:green', lw = '3', label = 'compute node, max mem')
#
# interactive 

max_memory, max_walltime, max_cpus, max_sockets, max_cores, max_threads  = mem_usage.max_resources_per_node("interactive")
max_memory_interactive = max_memory

mem_size, mem_unit = mem_usage.exctract_mem_size_and_unit(max_memory_interactive)
mem_converter = mem_usage.mem_unit_converter(old_unit=mem_unit, new_unit='G')
ax2.axhline(mem_converter*mem_size, color = 'tab:red', lw = 3,  label = 'interactv node, max mem')

ax.set_title('Memory of MSWEP data', fontsize=14)
ax.grid(True)
ax2.grid(True)

ax2.legend(fontsize = 14)
ax.legend(loc='center left', fontsize = 14)
```
