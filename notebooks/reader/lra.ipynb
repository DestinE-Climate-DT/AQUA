## Low Resolution Archive (LRA)

This shows how to access and play with the LRA and can be used as playground to check that all variables produced are correct.
We can call the LRA by selecting as sources `lra-r100-mon` and in the future also `lra-r100-day`. `r100` is the current LRA resolution and it is 1deg. 

Here below you can see an example from IFS and FESOM. No fixer is needed since variables have been decumulated and corrected according to the AQUA data format. 
However, some inconsistency might still be present. 

```python
from aqua import Reader
reader = Reader(model="FESOM", exp="tco2559-ng5", source="lra-r100-mon")
data = reader.retrieve(fix=False)
data
```

```python
data['sst'][6,:,:].plot()
```

```python
reader = Reader(model="IFS", exp="tco2559-ng5", source="lra-r100-mon")
data = reader.retrieve(fix=False)
data
```

```python
data['tprate'][1,:,:].plot()
```
